```python
# Load datasets
d_fake = pd.read_csv('Fake.csv')
d_true = pd.read_csv('True.csv')
```

```python
# Assign classes to the datasets
d_fake["class"] = 0  # Fake news
d_true["class"] = 1  # Real news
```

```python
# Separate some data for manual testing
d_fake_manual_testing = d_fake.tail(10).copy()
d_true_manual_testing = d_true.tail(10).copy()
```

```python
# Drop the last 10 entries from each dataset for training
d_fake = d_fake.iloc[:-10]
d_true = d_true.iloc[:-10]
```

```python
# Assign labels to manual testing data
d_fake_manual_testing['class'] = 0
d_true_manual_testing['class'] = 1
```

```python

# Merge both datasets
data = pd.concat([d_fake, d_true], axis=0).reset_index(drop=True)
data = data.drop(['title', 'subject', 'date'], axis=1)
```

```python

# Text Preprocessing Function
def wordopt(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)  # Remove text within square brackets
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Remove URLs
    text = re.sub(r'<.*?>+', '', text)  # Remove HTML tags
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # Remove punctuation
    text = re.sub(r'\n', '', text)  # Remove newline characters
    text = re.sub(r'\w*\d\w*', '', text)  # Remove words containing numbers
    return text
```

```python
# Apply preprocessing to the text column
data['text'] = data['text'].apply(wordopt)
```

```python
# Separate features and target
X = data['text']
y = data['class']
```

```python
# Handle class imbalance using RandomOverSampler
from imblearn.over_sampling import RandomOverSampler
oversample = RandomOverSampler()
X_resampled, y_resampled = oversample.fit_resample(pd.DataFrame(X), y)
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X_resampled.iloc[:, 0], y_resampled, test_size=0.25, random_state=42)
```

```python
# Convert text data into TF-IDF vectors
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
# Logistic Regression Model with class balancing
LR = LogisticRegression(class_weight='balanced', max_iter=1000)
LR.fit(xv_train, y_train)
```

```python
# Evaluate Logistic Regression Model
pred_lr = LR.predict(xv_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, pred_lr))
print(classification_report(y_test, pred_lr))
```

```python
# Decision Tree Model with tuned parameters to prevent overfitting
DT = DecisionTreeClassifier(max_depth=10, min_samples_split=5)
DT.fit(xv_train, y_train)
```

```python
# Evaluate Decision Tree Model
pred_dt = DT.predict(xv_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, pred_dt))
print(classification_report(y_test, pred_dt))
```

```python
# Function to return output label
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

# Manual Testing Function
def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_xv_test = vectorization.transform(new_def_test["text"])
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    
    print("\nLR Prediction: {} \nDT Prediction: {}".format(output_label(pred_LR[0]),
                                                           output_label(pred_DT[0])))
```

```python
# Test with a sample input
news = str(input("Enter the news text to check: "))
manual_testing(news)
```
